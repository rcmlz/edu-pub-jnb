## Daten analysieren

### Welche Klasse erreichte durchschnittlich die meisten/wenigsten Punkte

```python
df.pivot_table(values  = "Punkte",
               index   = ["Klasse"],
               aggfunc = ["mean"]).sort_values(by=("mean", "Punkte"))
```

```python
df.pivot_table(values  = "Punkte",
               index   = ["Jahr", "Klasse"],
               aggfunc = ["min", "median", "mean", "max"]).sort_values(by=("mean", "Punkte"))
```

## Continuous Chat

Priming

```python
priming = "You are the best Python coder in the world and know Pandas inside out. You give short and informative answers. You code like a student in 10th grade."
messages = [
    {"role": "system", "content": priming}
]
```

### First question

```python
prompt = "Why is Pandas better than Excel?"
```

Add prompt to conversation

```python
messages.append({"role": "user", "content": prompt})
```

Execute querry

```python
response = client.chat(model=model, messages=messages)
```

Add response to conversation -  so LLM remembers

```python
messages.append({"role": "assistant", "content": response.message.content})
```

Show answer

```python
Markdown(response.message.content)
```

### Follow up question 1
continuation on chat

```python
prompt = "Explain that in more detail"
```

```python
messages.append({"role": "user", "content": prompt})
response = client.chat(model=model, messages=messages)
messages.append({"role": "assistant", "content": response.message.content})
Markdown(response.message.content)
```

### 1) Einf√ºhrung

> Eine `while`-Schleife wiederholt einen Block so lange, wie eine Bedingung `True` ist.

**Syntax (kurz):**

`while <bedingung>:`

Schleifen sind n√ºtzlich, wenn die Anzahl der Wiederholungen nicht vorher bekannt ist.


````python
# Einfaches Beispiel: Z√§hle von 1 bis 5 mit einer while-Schleife
n = 1
while n <= 5:
    print('Durchlauf', n)
    n += 1

# Das ergibt 5 Zeilen. Diesen Code kannst du mit Run ‚Üí Run All Cells ausf√ºhren.
````

### 2) Abbruch mit `break`

Manchmal will man die Schleife vorzeitig beenden. Dazu verwendet man `break`.


````python
# Beispiel: Suche nach der ersten Zahl, die durch 7 teilbar ist
x = 1
while True:  # Endlosschleife mit Abbruch
    if x % 7 == 0:
        print('Erste durch 7 teilbare Zahl:', x)
        break
    x += 1

# Wichtig: Endlosschleifen immer so gestalten, dass sie sicher abbrechen.
````

### 3) Eingaben und Schleifen

Hier zeigen wir, wie man `input()` verwenden kann ‚Äî der Zellenaufruf ruft **nicht** `input()` automatisch beim `Run All` auf,
sondern ist als Funktion vorbereitet, damit `Run All Cells` nicht blockiert. Du kannst die Funktion einzeln ausf√ºhren, um interaktiv zu testen.

```python
# Beispiel (nicht automatisch aufgerufen):
# def interactive_counter():
#     limit = int(input('Gib eine obere Grenze ein: '))
#     i = 1
#     while i <= limit:
#         print(i)
#         i += 1
```


````python
def interactive_counter_simulated(limit=3):
    '''Simuliertes Beispiel, damit Run All Cells nicht blockiert.'''
    i = 1
    while i <= limit:
        print(i)
        i += 1

# Funktion aufrufen mit einer Beispiel-Grenze
interactive_counter_simulated(3)
````

### 4) Aufgaben / kleine √úbungen

1. Schreibe eine `while`-Schleife, die die Quadratzahlen (1, 4, 9, ...) bis zu 100 ausgibt.
2. Schreibe eine `while`-Schleife, die die Benutzereingabe (wiederholte Strings) liest, bis der Benutzer `stop` eingibt. (Testweise als Funktion implementiert.)


````python
# Aufgabe 1: Quadratzahlen bis 100
n = 1
while n*n <= 100:
    print(n*n)
    n += 1

# Aufgabe 2: simulierte stop-Schleife (nicht interaktiv, damit Run All Cells durchl√§uft)
def simulated_input_loop(sequence):
    idx = 0
    while idx < len(sequence):
        s = sequence[idx]
        print('eingabe:', s)
        if s.lower() == 'stop':
            print('Stop erkannt ‚Äî Schleife beendet')
            break
        idx += 1

simulated_input_loop(['hallo', 'test', 'stop', 'noch'])
````
